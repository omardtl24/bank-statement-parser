# How to use Loaders

## PDF Loader

Pdf loader includes the capability of parsing PDFs with passwords

```python
import fitz
from IPython.display import display
from PIL import Image
import io

doc = fitz.open("testFile.pdf")

for page in doc:
    pix = page.get_pixmap()
    img = Image.open(io.BytesIO(pix.tobytes("png")))
    display(img)
```

```python
from bankparser.loader import PDFLoader

pdf_loader = PDFLoader()

normal_loader = pdf_loader.load("testFile.pdf")
protected_loader = pdf_loader.load("testFile_protected.pdf", password="1234")

print("Normal PDF:")
print(normal_loader)

print("\nProtected PDF:")
print(protected_loader)
```
